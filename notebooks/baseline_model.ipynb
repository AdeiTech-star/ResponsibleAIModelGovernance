```python
# Create a dictionary mapping the current names to your new names
rename_mapping = {
    'PAY_0': 'Sept_hist',
    'PAY_2': 'Aug_hist',
    'PAY_3': 'July_hist',
    'PAY_4': 'June_hist',
    'PAY_5': 'May_hist',
    'PAY_6': 'Apr_hist',

    'BILL_AMT1': 'Sept_bill',
    'BILL_AMT2': 'Aug_bill',
    'BILL_AMT3': 'July_bill',
    'BILL_AMT4': 'June_bill',
    'BILL_AMT5': 'May_bill',
    'BILL_AMT6': 'Apr_bill',

    'PAY_AMT1': 'Sept_pay',
    'PAY_AMT2': 'Aug_pay',
    'PAY_AMT3': 'July_pay',
    'PAY_AMT4': 'June_pay',
    'PAY_AMT5': 'May_pay',
    'PAY_AMT6': 'Apr_pay',

    'default payment next month' : 'default'
}

# Apply the rename function to your dataframe
data.rename(columns=rename_mapping, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(5,5))
sns.countplot(data = data, x='SEX')

plt.title("Gender distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

Gender (1 = male; 2 = female)

```python
plt.figure(figsize=(5,5))
sns.countplot(data = data, x='EDUCATION', palette='coolwarm')

plt.title("Education distribution")
plt.xlabel("Education Level")
plt.ylabel("Count")

plt.show()
```

Education (1 = graduate school; 2 = university; 3 = high school; 4 = others)

```python
plt.figure(figsize=(5,5))
sns.countplot(data = data, x='MARRIAGE')

plt.title("Marriage distribution")
plt.xlabel("Married")
plt.ylabel("Count")

plt.show()
```

```python
# Age distribution:

plt.figure(figsize=(12,8))
sns.countplot(data = data, x='AGE')

plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```

```python
print(f"Minimum age: {data['AGE'].min()}")
print(f"Maximum age: {data['AGE'].max()}")
print(f"Mean age: {data['AGE'].mean().round(0)}")
```
