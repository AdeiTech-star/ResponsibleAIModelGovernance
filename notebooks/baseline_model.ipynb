Mapping demographic labels

```python
data['SEX'] = data['SEX'].map({1:'Male', 2:'Female'})

data['EDUCATION'] = data['EDUCATION'].map(
    {
        1: 'Graduate',
        2: 'University',
        3: 'HighSchool',
        4: 'Other',
        5: 'Other',
        6: 'Other'
    }
)

data['MARRIAGE'] = data['MARRIAGE'].map(
    {
        1: 'Married',
        2: 'Single',
        3: 'Other',
        0: 'Other'
    }
)
```

```python
#Creating age groups

data['AGE_GROUP'] = pd.cut(
    data['AGE'], bins=[20,30,45,80],
    labels=['Under 30', '30-45','45+']
)
```

```python
#Now we can check the default rate depending on the demographics, like your gender, educational level etc

demographics = ['SEX', 'EDUCATION', 'AGE_GROUP', 'MARRIAGE'] #these are the columns i'll be focusing on

for col in demographics:
    ct = pd.crosstab(data[col], data['default'], normalize='index')
    ct.columns = ['No default', 'Default']
    print(f"The rate of default by: {col}")
    print((ct.round(2)) * 100)
```

```python
# visualising the grouped data

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i,  col in enumerate(demographics):
    ct = pd.crosstab(data[col], data['default'], normalize='index')
    ct.columns = ['No default', 'Default']
    ct['Default'].plot(kind='bar', ax=axes[i], color='blue', edgecolor='black')
    axes[i].set_title(f'Default Rate by {col}', fontsize=13, fontweight='bold')
    axes[i].set_ylabel('Default Rate')
    axes[i].set_ylim(0, 0.4)
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].axhline(y=data['default'].mean(), color='gray', linestyle='--', label='Overall avg')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
#Now we do an intersectional analysis

print("=====What's the relationship between SEX  and EDUCATION ======")
ct1 = data.groupby(['SEX', 'EDUCATION'])['default'].mean().unstack()
print(f"{(ct1 * 100).round(2)}")


print("=====What's the relationship between SEX  and AGE_GROUP ======")
ct2 = data.groupby(['SEX', 'AGE_GROUP'])['default'].mean().unstack()
print(f"{(ct2 * 100).round(2)}")

print("=====What's the relationship between MARRIAGE and AGE_GROUP ======")
ct3 = data.groupby(['MARRIAGE', 'AGE_GROUP'])['default'].mean().unstack()
print(f"{(ct3 * 100).round(2)}")
```

Here are the observations:

1. Overrall, men have the highest default which could reflect differences in credit behaviour, amounts and other factors not campured.
2. With respect to education, those with just High School have a high default rate. This could mean they maybe aren't very well informomed with money, and we also see that those with Other(which also has a very small sample size, not sure we could get insight from it), and graduate level, have low default rate. So this could potentially mean, the more you're educated, the better the relationship with money, and you probably have a good job.
3. People who are married have a higher default rate. This could be due to the influence from the partner in order to provide, thus taking more loans than you could afford.
4. People older than 45 are the ones with the highest default rate of 25%. But it's not so much far from the other ages.

5. The relationship between gender and education, we see that men with graduate level education have the highest default as compared to women, which also takes us back to men being risk takers.
